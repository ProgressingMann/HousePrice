```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].apply(lambda x: x.year)
```

```python
df.corr()['price']
```

### Removing Unnecessary Features.

```python
input_features = df.corr()[(df.corr()['price'] > 0.1).values].index
```

```python
df = df[input_features]
```

```python
df
```

### Normalizing the data

```python
StandardScaler().fit(df[['bedrooms', 'sqft_living']])
```

```python
def Normalization(df):
    df = df[input_features]
    std_scaler = StandardScaler().fit(df)
    normalized_df = std_scaler.transform(df)
    
    return normalized_df, std_scaler
```
